**2. EXPLORATORY DATA ANALYSIS**

According to Allen et al. (2010), there are several key factors of attrition or turnovers:
* Job characteristics, such as payment. Employees tend to leave their jobs to take higher paying jobs somewhere else.
* Job environment: Conflict, participation in decision making, and environmental fit. Relationship with supervisor has a big role in turnover decision, and employees who have many connections to the organization are less likely to leave.
* Job dissatisfaction: It is one of the most consistent cause of turnover, and employees who are compatible with his job are more likely to stay.

For our Exploratory Data Analysis, let's compare Attrition to columns based on the factors mentioned before:
* Job Characteristic: Column **MonthlyIncome**
* Job Environment: Column **RelationshipSatisfaction**, **JobInvolvement**, and **EnvironmentalSatisfaction**
* Job Dissatisfaction: Column **JobSatisfaction**

**2.1.** Attrition and Payment

```python
import seaborn as sns
```

```python
dfn.head(5)
```

```python
## Viz Title
plt.title('Monthly Income and Attrition')

## Visualization
sns.histplot(x ='MonthlyIncome',
             hue ='Attrition',
             data = dfn,
             kde = True,
             palette = 'mako')

## Label Settings
plt.xlabel('Monthly Income')
plt.ylabel('Number of Employee')
```

Based on visualization, overall **employess that has attrited were mostly people with a monthly income of around $2500**. To take a further look, we need to find what exactly determines the monthly income.

For this reason, we are going to look at correlations between Monthly Income and other columns.

```python
# Select only columns with int data types
df_num = dfn.select_dtypes(include='number')

# Make correlation Table
df_corr = df_num.corr()

df_corr
```

```python
# Visualize correlation of the data using Heatmap
plt.figure(figsize=(15, 15))
plt.title('Correlation')
sns.heatmap(df_corr,
            annot=True,
            fmt='.2f')
```

Based on the heatmap above, Montly Income has a relatively high correlation with Job Level (0.95) and Total Working Years (0.77). For this reason, let's take a deeper dive by comparing Monthly Income to Job Level and Total Working Years, filtered by only people who attrited.

**2.1.1** Monthly Income and Job Level (Attrited only)

```python
## Filter table only people who attrited
dfn_attr = dfn[dfn['Attrition'].isin(['Yes'])]

## Viz Title
plt.title('Monthly Income and Job Level')

## Visualization
sns.histplot(x ='MonthlyIncome',
             hue ='JobLevel',
             data = dfn_attr,
             kde = True,
             palette = 'mako')

## Label Settings
plt.xlabel('Monthly Income')
plt.ylabel('Job Level')
```

In this data, higher Job Level means higher position in the company.

From Monthly Income and Job Level visualization, **people with monthly income of around $2500 are mostly of lower job levels, specifically level 1**. There are some level 2 jobs, but the numbers are not significant. 

**2.1.2** Monthly Income and  Total Working Years (Attrited only)

```python
## Filter table only people who attrited
dfn_attr = dfn[dfn['Attrition'].isin(['Yes'])]

## Viz Title
plt.title('Monthly Income and Total Working Years')

## Visualization
sns.lineplot(x = 'TotalWorkingYears', y = 'MonthlyIncome', data = dfn_attr)

## Label Settings
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
```

**2.2.1.** Attrition and Relationship Satisfaction

```python
# Viz title
plt.title('Attrition and Relationship Satisfaction')

# Visualization
sns.countplot(x = 'RelationshipSatisfaction',
              hue = 'Attrition',
              data = dfn,
              palette = 'mako')

# Viz settings
plt.xlabel('Relationship Satisfaction')
plt.ylabel('Attrition')
```

Based on the chart above, more people leave the company when they are more satisfied. However, attrited number from low relationship satisfaction (1) is almost as many as higher relationship satisfaction (3) despite the latter having much more people who stays. Therefore, the percentage of level 1 relationship satisfaction should be hight

To find out the percentage, we need to calculate the percentage and make a new visualization.

```python
# Group by RelationshipSatisfaction and Attrition
grouped = dfn.groupby(['RelationshipSatisfaction', 'Attrition']).size().unstack(fill_value=0)

# Calculate the percentage of 'Yes' attrition
grouped['Yes_Percentage'] = (grouped['Yes'] / (grouped['Yes'] + grouped['No'])) * 100

# Display the results
grouped
```

```python
# Create a bar plot
grouped['Yes_Percentage'].plot(kind='bar', color='skyblue')

# Add title and labels
plt.title('Attrition Percentage by Relationship Satisfaction')
plt.xlabel('Relationship Satisfaction')
plt.ylabel('Attrition Percentage')

# Show the plot
plt.show()
```

Based on the percentage, **people with very low relationship satisfaction tend to leave the company.**

**3. INSIGHT & CONCLUSION**

* **Low-Paid Employees Are Leaving**: Employees are more likely to quit if they earn between $2,500 and $5,000. They are typically in lower job levels with less than five years of experience.
* **Unhappy with Relationships and Environment**: Even though fewer employees fall into this group, employees with very low satisfaction in relationships and their work environment are leaving at higher rates (attrition rate over 20%) compared to those with better satisfaction (attrition rate 13-15%).
* **Less Involved Employees Are Leaving**: Employees who feel less involved in their jobs have a higher turnover rate (21.5%) than those who feel more involved (15.7%).
* **Dissatisfied Employees Are Quitting**: Employees with very low job satisfaction are more likely to leave (attrition rate over 20%), compared to those who are more satisfied (attrition rate around 10%).
